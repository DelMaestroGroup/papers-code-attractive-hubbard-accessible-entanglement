```julia
using CSV, Plots, LaTeXStrings
cur_colors = get_color_palette(:auto, plot_color(:white));
```

```julia
csv_reader = CSV.File("1d_SSHH_U4.0_L12_nup6_ndn6_proj_OBC.csv", header=0)
l = length(csv_reader)
dt_list = -0.5:0.05:0.5

S2m_U4_OBC = zeros(Float64, 15, l)
S2n_U4_OBC = zeros(Float64, 15, l)

for i in 1:length(csv_reader)
    @views copyto!(S2m_U4_OBC[:, i], parse.(Float64, split(csv_reader[i][1], r"\[|\s|\n|\]", keepempty=false)))
    @views copyto!(S2n_U4_OBC[:, i], parse.(Float64, split(csv_reader[i][2], r"\[|\s|\n|\]", keepempty=false)))
end
```

```julia
csv_reader = CSV.File("1d_SSHH_U4.0_L12_nup6_ndn6_proj_PBC.csv", header=0)
l = length(csv_reader)
dt_list = -0.5:0.05:0.5

S2m_U4_PBC = zeros(Float64, 15, l)
S2n_U4_PBC = zeros(Float64, 15, l)

for i in 1:length(csv_reader)
    @views copyto!(S2m_U4_PBC[:, i], parse.(Float64, split(csv_reader[i][1], r"\[|\s|\n|\]", keepempty=false)))
    @views copyto!(S2n_U4_PBC[:, i], parse.(Float64, split(csv_reader[i][2], r"\[|\s|\n|\]", keepempty=false)))
end
```

```julia
plot(
    dt_list, S2n_U4_OBC[8, :],
    label = L"n = 7",
    marker=(4, :square),
    xlabel=L"\delta t", ylabel=L"S^{(2)}_{A_n}",
    title="half-filled, OBC, U=4, L=12",
    framestyle=:box, grid=:false,
    fontfamily="Computer Modern"
)

plot!(
    dt_list, S2n_U4_OBC[7, :],
    label = L"n = 6",
    marker=(5, :circle)
)

plot!(
    dt_list, S2n_U4_OBC[6, :],
    label = L"n = 5",
    marker=(4, :square)
)

plot!(
    dt_list, S2n_U4_OBC[5, :],
    label = L"n = 4",
    marker=(4, :square)
)

plot!(
    dt_list, S2n_U4_OBC[9, :],
    label = L"n = 8",
    marker=(4, :square)
)
```

```julia
plot(
    dt_list, vcat(S2m_U4_OBC[8, 1:10], S2m_U4_OBC[7, 11:end]),
    label = L"m = 0",
    marker=(4, :square),
    xlabel=L"\delta t", ylabel=L"S^{(2)}_{A_m}",
    title="half-filled, OBC, U=4, L=12",
    framestyle=:box, grid=:false,
    fontfamily="Computer Modern"
)

plot!(
    dt_list, vcat(S2m_U4_OBC[7, 1:10], S2m_U4_OBC[6, 11:end]),
    label = L"m = -1",
    marker=(4, :square)
)

plot!(
    dt_list, vcat(S2m_U4_OBC[9, 1:10], S2m_U4_OBC[8, 11:end]),
    label = L"m = +1",
    marker=(5, :circle)
)

plot!(
    dt_list, vcat(S2m_U4_OBC[6, 1:10], S2m_U4_OBC[5, 11:end]),
    label = L"m = -2",
    marker=(4, :square)
)

plot!(
    dt_list, vcat(S2m_U4_OBC[10, 1:10], S2m_U4_OBC[9, 11:end]),
    label = L"m = +2",
    marker=(5, :circle)
)
```

```julia
plot(
    dt_list, S2n_U4_OBC[8, :] - S2n_U4_PBC[8, :]/2,
    label = L"n = 7",
    marker=(4, :square),
    xlabel=L"\delta t", ylabel=L"S^{(2)}_{A_n}",
    title="half-filled, OBC, U=4, L=12",
    framestyle=:box, grid=:false,
    fontfamily="Computer Modern"
)

plot(
    dt_list, S2n_U4_OBC[7, :] - S2n_U4_PBC[7, :]/2,
    label = L"n = 6",
    marker=(4, :square),
    xlabel=L"\delta t", ylabel=L"S^{(2)}_{A_n}",
    title="half-filled, OBC, U=4, L=12",
    framestyle=:box, grid=:false,
    fontfamily="Computer Modern"
)
```
