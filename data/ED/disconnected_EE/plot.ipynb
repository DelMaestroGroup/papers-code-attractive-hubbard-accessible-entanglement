```julia
using CSV, Plots, LaTeXStrings
```

```julia
csv_reader = CSV.File("./1d_SSHH_U4.0_L12_nup6_ndn6_A.csv", header=0)
l = length(csv_reader)
dt_list = -0.5:0.05:0.5

S2_U4_A = zeros(Float64, l)
S2m_U4_A = zeros(Float64, l)
S2n_U4_A = zeros(Float64, l)

Pm_U4_A = zeros(Float64, 13, l)
Pn_U4_A = zeros(Float64, 13, l)
P2m_U4_A = zeros(Float64, 13, l)
P2n_U4_A = zeros(Float64, 13, l)

for i in 1:length(csv_reader)
    S2_U4_A[i] = csv_reader[i][1]
    S2m_U4_A[i] = csv_reader[i][2]
    S2n_U4_A[i] = csv_reader[i][3]

    P2m_U4_A[:, i] = parse.(Float64, split(csv_reader[i][5], r"\[|\s|\n|\]", keepempty=false))
    P2n_U4_A[:, i] = parse.(Float64, split(csv_reader[i][6], r"\[|\s|\n|\]", keepempty=false))

    Pm_U4_A[:, i] = parse.(Float64, split(csv_reader[i][8], r"\[|\s|\n|\]", keepempty=false))
    Pn_U4_A[:, i] = parse.(Float64, split(csv_reader[i][9], r"\[|\s|\n|\]", keepempty=false))
end
```

```julia
csv_reader = CSV.File("./1d_SSHH_U4.0_L12_nup6_ndn6_AcapB.csv", header=0)
l = length(csv_reader)
dt_list = -0.5:0.05:0.5

S2_U4_AcapB = zeros(Float64, l)
S2m_U4_AcapB = zeros(Float64, l)
S2n_U4_AcapB = zeros(Float64, l)

Pm_U4_AcapB = zeros(Float64, 15, l)
Pn_U4_AcapB = zeros(Float64, 15, l)
P2m_U4_AcapB = zeros(Float64, 15, l)
P2n_U4_AcapB = zeros(Float64, 15, l)

for i in 1:length(csv_reader)
    S2_U4_AcapB[i] = csv_reader[i][1]
    S2m_U4_AcapB[i] = csv_reader[i][2]
    S2n_U4_AcapB[i] = csv_reader[i][3]

    @views copyto!(P2m_U4_AcapB[:, i], parse.(Float64, split(csv_reader[i][5], r"\[|\s|\n|\]", keepempty=false)))
    @views copyto!(P2n_U4_AcapB[:, i], parse.(Float64, split(csv_reader[i][6], r"\[|\s|\n|\]", keepempty=false)))

    @views copyto!(Pm_U4_AcapB[:, i], parse.(Float64, split(csv_reader[i][8], r"\[|\s|\n|\]", keepempty=false)))
    @views copyto!(Pn_U4_AcapB[:, i], parse.(Float64, split(csv_reader[i][9], r"\[|\s|\n|\]", keepempty=false)))
end
```

```julia
csv_reader = CSV.File("./1d_SSHH_U4.0_L12_nup6_ndn6_AcupB.csv", header=0)
l = length(csv_reader)
dt_list = -0.5:0.05:0.5

S2_U4_AcupB = zeros(Float64, l)
S2m_U4_AcupB = zeros(Float64, l)
S2n_U4_AcupB = zeros(Float64, l)

Pm_U4_AcupB = zeros(Float64, 15, l)
Pn_U4_AcupB = zeros(Float64, 15, l)
P2m_U4_AcupB = zeros(Float64, 15, l)
P2n_U4_AcupB = zeros(Float64, 15, l)

for i in 1:length(csv_reader)
    S2_U4_AcupB[i] = csv_reader[i][1]
    S2m_U4_AcupB[i] = csv_reader[i][2]
    S2n_U4_AcupB[i] = csv_reader[i][3]

    @views copyto!(P2m_U4_AcupB[:, i], parse.(Float64, split(csv_reader[i][5], r"\[|\s|\n|\]", keepempty=false)))
    @views copyto!(P2n_U4_AcupB[:, i], parse.(Float64, split(csv_reader[i][6], r"\[|\s|\n|\]", keepempty=false)))

    @views copyto!(Pm_U4_AcupB[:, i], parse.(Float64, split(csv_reader[i][8], r"\[|\s|\n|\]", keepempty=false)))
    @views copyto!(Pn_U4_AcupB[:, i], parse.(Float64, split(csv_reader[i][9], r"\[|\s|\n|\]", keepempty=false)))
end
```

```julia
plot(
    dt_list, (S2_U4_A + S2_U4_B - S2_U4_AcupB - S2_U4_AcapB) / log(2),
    label = L"S^{\mathrm{D}}_2 / \ln2",
    marker=(4, :square),
    xlabel=L"\delta t", ylabel=L"S^D_2",
    title="half-filled, disconnected, U=4, L=12",
    framestyle=:box, grid=:false,
    fontfamily="Computer Modern"
)

plot!(
    -0.5:0.05:0.5, (S2m_U4_A + S2m_U4_B - S2m_U4_AcupB - S2m_U4_AcapB) / log(2),
    label = L"S_{2,m}^{\mathrm{D, acc}} / \ln2",
    marker=(4, :square)
)

plot!(
    -0.5:0.05:0.5, (S2n_U4_A + S2n_U4_B - S2n_U4_AcupB - S2n_U4_AcapB) / log(2),
    label = L"S_{2,n}^{\mathrm{D, acc}} / \ln2",
    marker=(4, :square)
)
```
