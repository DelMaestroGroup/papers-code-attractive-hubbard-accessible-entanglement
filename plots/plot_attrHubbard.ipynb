```julia
using DelimitedFiles, Plots, LaTeXStrings
cur_colors = get_color_palette(:auto, plot_color(:white));
```

### 1D attractive Hubbard model

```julia
data = readdlm("../data/ED/attrHubbard/L8_nUp4_nDn4.txt");
U = data[1:3:end, 1];
S2 = data[1:3:end, 2];
S2m = data[1:3:end, 4]
S2n = data[1:3:end, 5];

plt1 = plot(
    U, S2,
    label=L"S_2",
    xlabel = L"U", ylabel=L"S_2",
    lc=cur_colors[1], mc=cur_colors[1],
    marker=(:square, 4),
    fontfamily="Computer Modern",
    ylim=(0.0,2.0),
    #legendfontsize = 12,
    #guidefontsize = 11,
    #tickfontsize = 10,
    framestyle=:box, grid=:false,
    legend=:topleft,
    #title=L"L=4\times 2, n_{\uparrow} = n_{\downarrow} = 4"
)

plot!(
    U, S2m,
    label=L"S^{\mathrm{acc}}_{2,m}",
    lc=cur_colors[2], mc=cur_colors[2],
    marker=(:square, 4)
)

plot!(
    U, S2n,
    label=L"S^{\mathrm{acc}}_{2,n}",
    lc=cur_colors[3], mc=cur_colors[3],
    marker=(:square, 4)
)
```

```julia
savefig("../figures/EE_attrHubbard_1d_Ne8.pdf")
savefig("../figures/EE_attrHubbard_1d_Ne8.svg")
savefig("../figures/EE_attrHubbard_1d_Ne8.png")
```
