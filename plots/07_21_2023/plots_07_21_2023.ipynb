```julia
data = load("../../data/07_21_2023/H_GS_N64_Lx8_Ly8_LA32_beta18.0.jld")

plot(
    data["U"], data["Hm_avg"], yerror=data["Hm_err"],
    marker=(4, :square),
    xlabel=L"U", ylabel=L"H_{1}",
    label=L"H_{1, m}",
    fontfamily="Computer Modern",
    legendfontsize = 12,
    guidefontsize = 12,
    tickfontsize = 10,
    framestyle=:box, grid=:false,
    legend=:bottomright,
    foreground_color_legend = nothing
)

plot!(
    data["U"], data["Hn_avg"], yerror=data["Hn_err"],
    marker=(5, :circle),
    label=L"H_{1, n}"
)
```

```julia
data = load("../../data/07_21_2023/Pn_GS_N64_Lx8_Ly8_LA32_beta18.0.jld")
data1 = load("../../data/07_21_2023/Pn_FT_N64_Lx8_Ly8_LA32_beta5.0.jld")
data2 = load("../../data/07_21_2023/Pn_FT_N64_Lx8_Ly8_LA32_beta8.0.jld")

plt_Pn = []
for i in 1:8
    plt = plot(
        data["Pn_avg"][:,i], yerror=data["Pn_err"][:, i],
        marker=(6, :circle),
        ylim=(0,0.5),
        label=L"T=0", ylabel=L"P_n",
        fontfamily="Computer Modern",
        legendfontsize = 12,
        guidefontsize = 12,
        tickfontsize = 10,
        framestyle=:box, grid=:false,
        legend=:topleft,
        foreground_color_legend = nothing
    )

    plot!(
        data1["Pn_avg"][:,i], yerror=data1["Pn_err"][:, i],
        marker=(4, :square),
        label=L"\beta=5.0"
    )

    plot!(
        data2["Pn_avg"][:,i], yerror=data2["Pn_err"][:, i],
        marker=(4, :square),
        label=L"\beta=8.0"
    )

    U = data["U"][i]
    annotate!(plt, 50, 0.25, text(L"U=%$U", 12))

    push!(plt_Pn, plt)
end
```
