```julia
δr = [(δrx, δry) for δrx in 0:4 for δry in 0:4]
q1 = (0, 0)
q2 = (π, π)
Pq = Measurement{Float64}[]
nq = Measurement{Float64}[]
Cq = Measurement{Float64}[]

for i in 1:8
    Ps = measurement.(data[i]["Ps_avg"], data[i]["Ps_err"])
    nᵢnⱼ = measurement.(data[i]["ninj_updn_avg"], data[i]["ninj_updn_err"])
    push!(Pq, real(sum([exp(dot(q1, δr[j])im) * Ps[j] for j in 1:25]))/25)
    push!(nq, real(sum([exp(dot(q2, δr[j])im) * nᵢnⱼ[j] for j in 1:25]))/25)
    nᵢ₊ᵣnᵢ = measurement.(data_corr[i]["ninj_avg"], data_corr[i]["ninj_err"])
    push!(Cq, real(sum([exp(dot(q2, δr[j])im) * nᵢ₊ᵣnᵢ[j] for j in 1:25]))/25)
end
```

```julia
plot(
    -1.0:-1.0:-8.0, Pq,
    xlabel=L"U", label=L"\tilde{P}_{\zeta}(0,0)",
    marker=(6, :diamond),
    fontfamily="Computer Modern",
    legendfontsize = 14,
    guidefontsize = 12,
    tickfontsize = 10,
    framestyle=:box, grid=:false,
    legend=:topright,
    foreground_color_legend = nothing
)

plot!(
    -1.0:-1.0:-8.0, Cq,
    label=L"\mathcal{C}(\pi,\pi)",
    marker=(5, :utriangle)
)

plot!(
    -1.0:-1.0:-8.0, nq,
    label=L"\mathcal{N}(\pi,\pi)",
    marker=(5, :square)
)
```
