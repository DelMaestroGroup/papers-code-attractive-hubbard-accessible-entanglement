```julia
using JLD, Plots, LaTeXStrings, Statistics
```

```julia
data = load("../data/QMC/num_U1.0_dt-0.4_beta36.0_5700.jld")
data2 = load("../data/QMC/denom_U1.0_dt-0.4_beta36.0_5700.jld");
```

```julia
scatter(
    data["expS2"], 
    label=false,
    xticks=false, xlabel=L"(\varphi_1, \varphi_2)",
    yticks=([1e-6, 1e-4, 1e-2, 1e0, 1e2, 1e4]),
    yaxis=:log10,
    ylabel=L"\det g_A^{\varphi_1, \varphi_2}",
    fontfamily="Computer Modern",
    legendfontsize = 10,
    guidefontsize = 11,
    tickfontsize = 10,
    framestyle=:box, grid=:false
)

hline!([mean(data["expS2"])], line=(2, :dash, :red), label="")
```

```julia
savefig("../figures/QMCsamples_Grover.svg")
savefig("../figures/QMCsamples_Grover.pdf")
savefig("../figures/QMCsamples_Grover.png")
```

```julia
cur_colors = get_color_palette(:auto, plot_color(:white))

p1 = scatter(
    data["p"], 
    label=false,
    xticks=false, xlabel=L"(\varphi_1, \varphi_2)",
    ylabel=L"p_{Z^2 \rightarrow Z^{(2)}_A}",
    fontfamily="Computer Modern",
    legendfontsize = 10,
    guidefontsize = 11,
    tickfontsize = 10,
    framestyle=:box, grid=:false
)

hline!(p1, [mean(data["p"])], line=(3, :dash, :red), label="")

p2 = scatter(
    data2["p"], 
    label=false,
    xticks=false, xlabel=L"(\varphi_1, \varphi_2)",
    ylabel=L"p_{Z^{(2)}_A \rightarrow Z^2}",
    fontfamily="Computer Modern",
    legendfontsize = 10,
    guidefontsize = 11,
    tickfontsize = 10,
    framestyle=:box, grid=:false
)

hline!(p2, [mean(data2["p"])], line=(3, :dash, :red), label="")

plot(p1, p2, layout=(2,1))
```
